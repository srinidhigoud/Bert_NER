```python
import json
import csv
```

```python
with open('training_dataset_2.json') as json_file:
    chatito_data = json.load(json_file)
    training_data = [['Sentence #','Word','Tag']]
    ct = 0
    for intent in chatito_data['intents'].keys():
        for query in chatito_data['intents'][intent]['utterances']:
            for text in query['data']:
                tag = 'O'
                if 'entity' in text.keys():
                    tag = text['entity']
                words = text['text'].split()
                for word in words:
                    training_data.append(['Sentence: {}'.format(ct), word, tag])
            training_data.append(['Sentence: {}'.format(ct), '.', 'O'])
            ct += 1
```

```python
training_data
```

```python
with open('training_data.csv', 'w') as writeFile:
    writer = csv.writer(writeFile)
    writer.writerows(training_data)
writeFile.close()
```

```python
with open('training_data.csv','r') as csvin, open('training_data.tsv', 'w') as tsvout:
    csvin = csv.reader(csvin)
    tsvout = csv.writer(tsvout, delimiter='\t')

    for row in csvin:
        tsvout.writerow(row)
```
